```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
import fitz  # pip install pymupdf

import os

import re

import pandas as pd



def parse_invoice_text(FILE_PATH):
    doc = fitz.open("/content/taco_Tax_Invoice_408_09042025_250409FS29010820.pdf")

    full_text = ""

    for page in doc:

        full_text += page.get_text()



    # Extract store/brand

    brand = re.search(r"Restaurant / Store Name\s*:\s*(.+)", full_text)

    brand = brand.group(1).strip() if brand else ""



    # Extract city

    city = re.search(r"City\s*:\s*(.+)", full_text)

    city = city.group(1).strip() if city else ""



    # Extract address

    addr_match = re.search(r"Address\s*:\s*(.*?)City\s*:", full_text, re.DOTALL)

    location = addr_match.group(1).replace("\n", " ").strip() if addr_match else ""



    return brand, city, location



data = []

pdf_dir = "drive"  # change if needed



for file in os.listdir(pdf_dir):

    if file.endswith(".pdf"):

        path = os.path.join(pdf_dir, file)

        try:

            b, c, l = parse_invoice_text(path)

            data.append({"File": file, "Brand": b, "City": c, "Location": l})

        except:

            data.append({"File": file, "Brand": "", "City": "", "Location": ""})



df = pd.DataFrame(data)

df.to_excel("swiggy_invoice_extract.xlsx", index=False)
```
